```python
from rdkit import Chem
```

```python
mol = Chem.MolFromSmiles("C(=C1OCc2ccccc21)c1ccccc1")

mol = Chem.MolFromSmiles("c3ccc(Cc1occ2ccccc12)cc3")
```

```python
mol = Chem.AddHs(mol)
```
